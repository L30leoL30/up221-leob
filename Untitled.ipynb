```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# File paths
geojson_path = "California_Counties.geojson"
csv_path = "Merged_Industry_Data.csv"

# Load the county boundaries
counties = gpd.read_file(geojson_path)

# Load the employment data
df = pd.read_csv(csv_path)

# Standardize county names for merging
df['County'] = df['County'].str.strip()
counties['NAME'] = counties['NAME'].str.strip()

# Merge the data on county name
gdf = counties.merge(df, left_on='NAME', right_on='County', how='left')

# Ensure all counties are included, even those without employment data
gdf['Difference in Employment  2020-2024'] = gdf['Difference in Employment  2020-2024'].fillna(0)

# Define color scale limits
vmin, vmax = gdf['Difference in Employment  2020-2024'].min(), gdf['Difference in Employment  2020-2024'].max()

# Create figure and axis
fig, ax = plt.subplots(1, 1, figsize=(10, 12))
ax.set_title("Change in Employment in the Carwash Industry by County (2020-2024)", fontsize=14)

# Plot the counties with the employment change data
gdf.plot(column='Difference in Employment  2020-2024', cmap='RdYlGn', linewidth=0.8, edgecolor='black', ax=ax,
         legend=True, vmin=vmin, vmax=vmax)

# Move the legend below the map
leg = ax.get_legend()
leg.set_bbox_to_anchor((0.5, -0.1))  # Adjust position

# Remove axis labels
ax.set_xticks([])
ax.set_yticks([])
ax.set_frame_on(False)

# Show the map
plt.show()
```
